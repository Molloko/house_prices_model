```python
# Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Preprocessing
from sklearn.impute import SimpleImputer

# Важная настройка для корректной настройки pipeline!
import sklearn
sklearn.set_config(transform_output="pandas")
```

```python
drop_features = ['BsmtHalfBath', 'BsmtFinSF2', 'MiscVal', 'LowQualFinSF', 'YrSold', 'OverallCond', 'MSSubClass', 'EnclosedPorch', 'KitchenAbvGr', 'Alley', 'PoolQC', 'Fence', 'MiscFeature', 'Utilities']

num_features = ['LotFrontage', 'GarageYrBlt']
cat_features = ['MasVnrType', 'FireplaceQu', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']

my_imputer = ColumnTransformer(
    transformers=[
        ('drop_features', 'drop', drop_features),
        ('num_imputer', SimpleImputer(strategy='mean'), num_features),
        ('cat_imputer', SimpleImputer(strategy="constant", fill_value="no_information"), cat_features)
    ],
    verbose_feature_names_out=False,
    remainder='passthrough'
)
```

```python
filled_data = my_imputer.fit_transform(X_train)
```

```python
pd.DataFrame(data={'NaN_count': filled_data.isna().sum(), 'unique_values': filled_data.nunique(), 'data_type':filled_data.dtypes})
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
```

```python
#бинарные и 3 колонки через OneHot
#остальные через Ordinal

cat_cols = X_train.select_dtypes(exclude="number").columns.tolist()

threshold = 3

cat_features_ohe = [c for c in cat_cols if X_train[c].nunique() <= threshold]
cat_features_ord = [c for c in cat_cols if X_train[c].nunique() > threshold]

my_encoder = ColumnTransformer(
    transformers=[
        ('cat_ohe', OneHotEncoder(sparse_output=False), cat_features_ohe),
        ('cat_ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_features_ord),
    ],
    verbose_feature_names_out = False,
    remainder = 'passthrough' 
)
```

```python
y_train_log = np.log1p(y_train)  # log(1 + y)
y_valid_log = np.log1p(y_valid)
```

```python
encoded_data = my_encoder.fit_transform(filled_data, y_train_log)

encoded_data.head()
```

```python
#нормализовать SF, Area, Porch, ID оставить как есть

num_std = ['LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'PoolArea', '3SsnPorch',
           'ScreenPorch']

my_scaler = ColumnTransformer(
    transformers=[
        ('std_scaler', StandardScaler(), num_std)
    ], 
    verbose_feature_names_out = False,
    remainder = 'passthrough'
)
```

```python
scaled_data = my_scaler.fit_transform(encoded_data, y_train_log)

scaled_data.head()
```

```python
# === 0) Импорты (если не были выше) ===
import numpy as np
import pandas as pd
from sklearn.base import clone
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error

# --- у тебя уже есть: imp ---
# imp = pd.Series(r.importances_mean, index=X_valid.columns).sort_values(ascending=False)

# === 1) Список к удалению и к оставлению ===
to_drop = imp[imp <= 0].index.tolist()
cols_keep = [c for c in X_train.columns if c not in to_drop]

# === 2) Числовые для скейлинга: урезаем твой список num_std ===
# num_std у тебя задан заранее
num_std_red = [c for c in num_std if c in cols_keep]

# === 3) Категориальные, проходящие через passthrough ===
other_cols = [c for c in cols_keep if c not in num_std_red]
cat_passthrough = [c for c in other_cols if not np.issubdtype(X_train[c].dtype, np.number)]

# трансформер: NaN -> "__MISSING__", всё в строки
# трансформер: NaN -> "__MISSING__", всё в строки (принимает DataFrame)
cat_fix = FunctionTransformer(
    lambda df: df.fillna('__MISSING__').astype(str),
    validate=False,
    feature_names_out='one-to-one'
)


# === 4) Новый препроцессор: scale числовые + привести категории к строкам, остальные — как есть ===
my_scaler_red = ColumnTransformer(
    transformers=[
        ('std_scaler', StandardScaler(), num_std_red),   # блок 1
        ('cat_fix',    cat_fix,         cat_passthrough) # блок 2 (категории)
    ],
    verbose_feature_names_out=False,
    remainder='passthrough'  # блок 3: прочие фичи без изменений
)

# индексы категориальных на ВЫХОДЕ препроцессора — это как раз блок 2 сразу после блокa 1
cat_idx_tr = list(range(len(num_std_red), len(num_std_red) + len(cat_passthrough)))

# === 5) Новый пайплайн: клонируем модель из старого пайпа ===
est = clone(ml_pipeline_cb.named_steps['model'])  # имя шага у тебя 'model'

ml_pipeline_cb_red = Pipeline([
    ('prep',  my_scaler_red),
    ('model', est)
])

# === 6) Обучение и предсказание на урезанных колонках ===
X_train_red = X_train[cols_keep]
X_valid_red = X_valid[cols_keep]

fit_kwargs = dict(model__verbose=0)
# только если есть категориальные после препроцессора — передаём их индексы
if len(cat_idx_tr) > 0:
    fit_kwargs['model__cat_features'] = cat_idx_tr

ml_pipeline_cb_red.fit(X_train_red, y_train_log, **fit_kwargs)
y_pred = ml_pipeline_cb_red.predict(X_valid_red)

# (опционально) метрики

rmse_orig =  np.sqrt(mean_squared_log_error(y_valid, np.expm1(y_pred)))

print(f"RMSE: {rmse_orig}")

# Далее для теста используем те же cols_keep:
# X_test_red = X_test[cols_keep]
# y_test_pred_log = ml_pipeline_cb_red.predict(X_test_red)
# y_test_pred = np.expm1(y_test_pred_log)
```
